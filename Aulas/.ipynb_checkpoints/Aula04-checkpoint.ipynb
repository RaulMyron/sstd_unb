```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

# Resposta ao impulso

Considere o sistema:

\begin{align}
	y\left[n\right] - 0.6 y\left[n-1\right] -0.16  y\left[n\right] = 5 x\left[n\right]
\end{align}

Determine a resposta ao impulso.

```python
#Crio as variáveis primeiro
n = np.arange(-2, 20, 1)
x = np.zeros_like(n, dtype=float)
x[n == 0] = 1

h = np.zeros_like(n, dtype=float)

for k in range(2,len(n)):
    h[k] = 0.16 * h[k-2] + 0.6 * h[k-1] + 5 * x[k]
    
print(f'n = {n}')
print(f'x = {x}')
print(f'h = {h}')

plt.stem(n, h)
```

```python
n = np.arange(0, 20, 1)
h0 = np.zeros_like(n, dtype=float)

h0 = 4 * (0.8) ** n + (-0.2)** n

h0 = np.concatenate((np.array([0,0],dtype=float), h0))
n = np.arange(-2, 20, 1)

plt.stem(n, h0)

print(f'Recursiva | Fechada')
for k in range(len(n)):
    print(f'{h[k]:.6f}  |  {h0[k]:.6f}')
```
