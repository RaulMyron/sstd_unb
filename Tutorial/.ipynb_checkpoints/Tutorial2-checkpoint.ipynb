```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
```

# Exemplo 5.10 do Livro

Considere a função de transferência:

$$\begin{align}
H\left(z\right) = \frac{z}{z - 0.8}
\end{align}$$

```python
#H(z) = z / (z - 0.8)
#Ou seja, b0 = 1, b1 = 0, a0 = 1, a1 = -0.8
b0 = 1
b1 = 0
a0 = 1
a1 = -0.8

w = np.linspace(-3 * np.pi, 3 * np.pi, 10000)
w_axis = np.exp(1j * w)

H = (b0 * w_axis + b1) / (a0 * w_axis + a1)
H_abs = abs(H)
H_phase_deg = np.angle(H, deg=True)

plt.plot(w, H_abs)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude')
plt.show()

plt.plot(w, H_phase_deg)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (Graus)')
plt.show()

#Note que é comum calcular a resposta em magnitude em dB e a resposta em fase em radianos
H_mag_db = 10 * np.log10(abs(H))
H_phase = np.angle(H)

plt.plot(w, H_mag_db)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude (dB)')
plt.show()

plt.plot(w, H_phase)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (Radianos)')
plt.show()
```

$$ H\left(z\right) = \frac{\left(z+1\right)\left(z-1\right)}{\left(z - \gamma e^{j\frac{\pi}{4}}\right)\left(z - \gamma e^{-j\frac{\pi}{4}}\right)  }$$

$$ H\left(e^{j\Omega}\right) = \frac{\left(e^{j\Omega}+1\right)\left(e^{j\Omega}-1\right)}{\left(e^{j\Omega} - \gamma e^{j\frac{\pi}{4}}\right)\left(e^{j\Omega} - \gamma e^{-j\frac{\pi}{4}}\right)  }$$


```python
gamma = 0.8
H1 = ((w_axis + 1)*(w_axis - 1))/((w_axis -gamma * np.exp(1j * np.pi / 4))*(w_axis -gamma * np.exp(-1j * np.pi / 4)))
H1_abs = abs(H1)
H1_phase_deg = np.angle(H1, deg=True)

plt.plot(w, H1_abs)
plt.title('Resposta em Frequência de H_1(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude')
plt.show()

plt.plot(w, H1_phase_deg)
plt.title('Resposta em Frequência de H_1(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (Graus)')
plt.show()
```
