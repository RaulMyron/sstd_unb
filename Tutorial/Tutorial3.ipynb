```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
import time
```

Considere uma senóide com frequência $f = 1$ Hz. Ou seja:

$$\begin{align}
x\left(t\right) &= \cos \omega t \\
                &= \cos 2 \pi f t \\
                &= \cos 2 \pi t
\end{align}$$

```python
t = np.linspace(0, 2, 1000)
x = np.cos(2 * np.pi * t)

plt.plot(t, x)
plt.title('Senóide de 1 Hz')
plt.xlabel('t (s)')
plt.ylabel('Magnitude')
plt.show()
```

Vamos amostrar o sinal $x\left(t\right)$ com uma taxa de amostragem $T = \frac{1}{6}$, ou seja:

$$
\begin{align}
x\left[ n \right] &= x\left(nT\right) \\
 &= cos \left( 2 \pi n T \right) \\
 &= cos \left( \frac{2 \pi} {6} n \right) \\
 &= cos \left(\Omega ~ n\right)
\end{align}
$$

Onde, naturalmente, $\Omega = \frac{2 \pi}{6}$.

Podemos observar o sinal amostrado:

```python
n = np.arange(0,13)
xa = np.cos (2 * np.pi / 6 * n)

plt.plot(t, x)
plt.stem(n/6,xa)
plt.title('Senóide de 1 Hz amostrada')
plt.xlabel('t (s)')
plt.ylabel('Magnitude')
plt.show()

print('As amostras de x são: ')
print(xa)
```

Considere agora o sinal $y\left(t\right)$, também uma senóide com frequência $f_2 = 7 Hz.$

$$\begin{align}
y\left(t\right) &= \cos \omega_2 t \\
                &= \cos 2 \pi f_2 t \\
                &= \cos 14 \pi t
\end{align}$$


```python
y = np.cos(2 * np.pi * 7 * t)

plt.plot(t, y)
plt.title('Senóide de 7 Hz')
plt.xlabel('t (s)')
plt.ylabel('Magnitude')
plt.show()
```

Vamos amostrar o sinal $y\left(t\right)$ com a mesma taxa de amostragem $T = \frac{1}{6}$, ou seja:

$$
\begin{align}
y\left[ n \right] &= y\left(nT\right) \\
 &= cos \left( 14 \pi n T \right) \\
 &= cos \left( \frac{14 \pi} {6} n \right) \\
 &= cos \left(\Omega_2 ~ n\right)
\end{align}
$$

Onde, naturalmente, $\Omega_2 = \frac{14 \pi}{6}$.

Podemos observar o sinal amostrado:

```python
ya = np.cos (2 * np.pi / 6 * n)

plt.plot(t, y)
plt.stem(n/6,ya)
plt.title('Senóide de 7 Hz amostrada')
plt.xlabel('t (s)')
plt.ylabel('Magnitude')
plt.show()

print('As amostras de y são: ')
print(ya)
```

Podemos plotar os dois sinais juntos para entender o que está acontencendo:

```python

plt.plot(t, x)
plt.plot(t,y)
plt.stem(n/6,xa)
plt.stem(n/6,ya)
plt.title('Senóides de 1 Hz e 7 Hz amostradas')
plt.xlabel('t (s)')
plt.ylabel('Magnitude')
plt.show()

print('As amostras de x são: ')
print(xa)
print('As amostras de y são: ')
print(ya)
```

Claramente as duas senóides no tempo, $x\left(t\right)$ e $y\left(t\right)$ são diferentes. Porém, a amostragem nesta taxa produz amostras iguais!

Você pode dizer que eu escolhi a dedo um sinal que geraria amostras iguais, o que é verdade. Mas o ponto é que sempre vai existir uma senóide com uma frequência tal que $\Omega_2 = \Omega_1 + 2 \pi m$, ou seja, sempre vai existir um _aliasing_. Além disso, como a transformada de fourier nos mostra que qualquer sinal pode ser representado por uma soma infinita de senóides, outros sinais sempre terão componentes nestas frequências de aliasing.
